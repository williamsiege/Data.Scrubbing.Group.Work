```python
# ============================================================================
# Knowledge Check 1 - Data Understanding Phase (CRISP-DM Model)
# ============================================================================
# Questions:
# 1. How many rows and columns are contained in the dataset?
# 2. What datatypes are contained in the dataset?
# 3. Is the dataset complete? i.e., no missing values?
# 4. List the number of Movies and TV Shows among the last 20 rows.
# ============================================================================
# In the context of Learning Agents and AI Data Mining, the Data Understanding
# phase is crucial. A learning agent uses its 'learning element' to acquire new
# knowledge from data, and its 'performance element' to act on that knowledge.
# The 'critic' component evaluates outcomes, while the 'problem generator'
# suggests new explorations. This Knowledge Check demonstrates the Data
# Understanding phase of the CRISP-DM methodology — where we identify, collect,
# and explore initial datasets to gain insights before data preparation and
# modelling.
# ============================================================================

import pandas as pd
import numpy as np

# --- Load the Netflix Titles dataset (Data Collection) ---
# A learning agent's first step is gathering data from its environment.
# Here, we read the CSV file into a pandas DataFrame for exploration.

df = pd.read_csv('netflix_titles.csv')

# ============================================================================
# Question 1: How many rows and columns are contained in the dataset?
# ============================================================================
# Using df.shape to retrieve the dimensions as a tuple (rows, columns).
# This helps a data mining agent understand the scale of the dataset it
# is working with — a key part of the Data Understanding phase.
rows, cols = df.shape
print("=" * 60)
print("Q1: Dataset Dimensions")
print("=" * 60)
print(f"The dataset contains {rows} rows and {cols} columns.")
print(f"Shape tuple: {df.shape}")
print()

# ============================================================================
# Question 2: What datatypes are contained in the dataset?
# ============================================================================
# Using df.dtypes to inspect the data type of each column.
# Understanding datatypes is essential for a learning agent's 'learning
# element' — it determines how the agent can process and analyse each
# feature (e.g., numeric vs categorical data require different treatment).
print("=" * 60)
print("Q2: Data Types in the Dataset")
print("=" * 60) # It creates one string made of 60 equals signs to visually separate the output sections.
print(df.dtypes)
print(f"\nUnique datatypes: {df.dtypes.unique().tolist()}")
print()

# ============================================================================
# Question 3: Is the dataset complete? i.e., no missing values?
# ============================================================================
# Using df.isnull().sum() to count missing values per column.
# Data completeness is critical in data mining — incomplete data can mislead
# a learning agent's 'performance element', leading to poor decisions.
# The 'critic' component would flag missing data as a quality issue that
# needs to be addressed in the Data Preparation phase.
missing_values = df.isnull().sum()
total_missing = np.sum(missing_values.values)

print("=" * 60) # It creates one string made of 60 equals signs to visually separate the output sections.
print("Q3: Data Completeness Check")
print("=" * 60)

if total_missing == 0:
    print("The dataset is COMPLETE — no missing values detected.")
else:
    print(f"The dataset is NOT complete. Total missing values: {total_missing}")
    print("\nMissing values per column:")
    # Display only columns that have missing values for clarity
    missing_only = missing_values[missing_values > 0]
    for column, count in missing_only.items():
        pct = (count / rows) * 100
        print(f"  - {column}: {count} missing ({pct:.2f}%)")

print()

# ============================================================================
# Question 4: List the number of Movies and TV Shows among the last 20 rows
# ============================================================================
# Using df.tail(20) to access the last 20 rows, then value_counts() on the
# 'type' column. This kind of exploratory analysis helps a learning agent's
# 'problem generator' understand the distribution of categories, which can
# inform future modeling decisions in the CRISP-DM pipeline.
last_20 = df.tail(20)
type_counts = last_20['type'].value_counts()

print("=" * 60)
print("Q4: Movies vs TV Shows in the Last 20 Rows")
print("=" * 60)
for content_type, count in type_counts.items():
    print(f"  - {content_type}: {count}")
print()

# --- Additional: Display full dataset info for verification ---
# df.info() provides a comprehensive summary combining answers to Q1-Q3,
# showcasing how a data mining agent consolidates its understanding.
print("=" * 60)
print("Full Dataset Info (df.info())")
print("=" * 60)
df.info()


```

```python
##### Knowledge Check 2 - Data Indexing.
# Based on the Netflix Titles Dataset, respond to these questions:
# 1. Give the number of all Movies originating from South Africa
# 2. What is the description of the movie in the 12th row?
# 3. From these South African movies, list only those released in 2021
# 4. List the entire cast of all movies released in 2021.

# ============================================================================
# Knowledge Check 2 - Data Indexing (CRISP-DM Model)
# ============================================================================
# In the Data Understanding phase of CRISP-DM, data indexing is a fundamental
# skill for AI data mining agents. It allows the agent to access specific
# subsets of data based on conditions, which is crucial for learning and
# problem-solving. This Knowledge Check focuses on indexing techniques to
# extract insights from the Netflix Titles dataset, demonstrating how a learning
# agent can navigate and manipulate data to answer specific questions.
# ============================================================================

# Importing necessary libraries for data manipulation and analysis.
import pandas as pd
import numpy as np

# --- Load the Netflix Titles dataset (Data Collection) ---
# This is the environment data our AI data mining agent learns from.
df = pd.read_csv('netflix_titles.csv')

# ============================================================================
# Q1: Number of Movies originating from South Africa
# ============================================================================
# Build a boolean mask:
# - type must be "Movie"
# - country contains "South Africa" (includes co-productions, e.g. "South Africa, UK")
sa_movie_mask = (
    (df['type'] == 'Movie')
    & (df['country'].fillna('').str.contains(r'\bSouth Africa\b', case=False, regex=True))
)

# Use numpy to sum True values in the boolean mask (True=1, False=0).
sa_movie_count = int(np.sum(sa_movie_mask))
south_africa_movies = df[sa_movie_mask]
print('=' * 60)
print('Q1: Number of Movies originating from South Africa')
print('=' * 60)
print(f"Total South African Movies: {sa_movie_count}")
print()

# ============================================================================
# Q2: Description of the movie in the 12th row
# ============================================================================
# DataFrame indexing is zero-based, so the 12th row is position 11.
row_12 = df.iloc[11]
print('=' * 60)
print('Q2: Description of the movie in the 12th row')
print('=' * 60)
print(f"Title: {row_12['title']}")
print(f"Description: {row_12['description']}")
print()

# ============================================================================
# Q3: South African movies released in 2021
# ============================================================================
# Reuse the South African movie subset from Q1 and filter release year.
sa_movies_2021 = south_africa_movies[south_africa_movies['release_year'] == 2021]
print('=' * 60)
print('Q3: South African Movies released in 2021')
print('=' * 60)
if sa_movies_2021.empty:
    print('No South African movies were released in 2021 in this dataset.')
else:
    print(sa_movies_2021[['title', 'release_year', 'country']].to_string(index=False))
print()

# ============================================================================
# Q4: Entire cast of all movies released in 2021
# ============================================================================
# Filter all movies released in 2021 (not limited to South Africa), then
# display each movie title with the full cast string.
movies_2021 = df[(df['type'] == 'Movie') & (df['release_year'] == 2021)].copy()
movies_2021['cast'] = movies_2021['cast'].fillna('Cast not listed')

print('=' * 60)
print('Q4: Entire cast of all movies released in 2021')
print('=' * 60)
if movies_2021.empty:
    print('No movies released in 2021 were found in this dataset.')
else:
    for _, movie in movies_2021[['title', 'cast']].iterrows():
        print(f"- {movie['title']}: {movie['cast']}")

# ============================================================================
# Why this matters for AI Data Mining Agents
# ============================================================================
# These indexing and filtering operations are foundational for AI agents:
# - They help the learning element focus on relevant subsets of data.
# - They improve the critic's ability to evaluate data quality and patterns.
# - They support problem generation by surfacing useful slices for deeper analysis.

```
